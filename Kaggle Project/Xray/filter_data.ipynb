```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_df = pd.read_csv('train.csv', encoding='utf-8')
```

```python
train_df
```

```python
have_lesion_df = train_df[train_df.class_id != 14].reset_index(drop = True)
have_lesion_df
```

```python
ver1_df = have_lesion_df.copy()
```

```python
ver1_df.shape
```

```python
# custom_dataset_annotation.ipynb, train.csv, (train, val directories) are located in same directories.
ver1_df['height'] = ver1_df.x_max - ver1_df.x_min
ver1_df['width'] = ver1_df.y_max - ver1_df.y_min
```

```python
import os

train_files = []
train_wh_files = []
val_files = []
val_wh_files = []

for dirname, _, filenames in os.walk('train'):
    for filename in filenames:
        train_files.append(filename[:-4])
        train_wh_files.append(filename)

for dirname, _, filenames in os.walk('val'):
    for filename in filenames:
        val_files.append(filename[:-4])
        val_wh_files.append(filename)
```

```python
len(train_files), len(val_files)
```

```python
train_files2 = np.sort(train_files)
train_files2
```

```python
loc_list = []
for item in ver1_df.image_id:
    if item in train_files2: # 30067
        loc_list.append('train/' + item + '.png')
    else: # 6029
        loc_list.append('val/' + item + '.png')
```

```python
ver1_df['location'] = loc_list
ver1_df
```

```python
# train, validation < split >
ver1_df.to_csv('ver1.csv', sep=',', na_rep='NaN')
```

```python
import cv2

# train_wh_files
with open('wh_train.csv', 'w') as f1:
    f1.write('id,width,height\n')
    for item in train_wh_files:
        image = cv2.imread(f'train/{item}', cv2.IMREAD_GRAYSCALE)
        w = image.shape[1]
        h = image.shape[0]
        f1.write(item[:-4] + ',' + str(w) + ',' + str(h) + '\n')
```

```python
# val_wh_files
with open('wh_val.csv', 'w') as f1:
    f1.write('id,width,height\n')
    for item in val_wh_files:
        image = cv2.imread(f'val/{item}', cv2.IMREAD_GRAYSCALE)
        w = image.shape[1]
        h = image.shape[0]
        f1.write(item[:-4] + ',' + str(w) + ',' + str(h) + '\n')
```
