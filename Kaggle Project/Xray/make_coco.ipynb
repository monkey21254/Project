```python
import json
from collections import OrderedDict
```

```python
images_list = []
for item in zip(train_df2.index.tolist(), train_df2.file_name, train_df2.width, train_df2.height):
    images_dict = dict(id = item[0], file_name = item[1] + '.png', width = item[2], height = item[3])
    images_list.append(images_dict)
    
images_list2 = []
for item in zip(val_df2.index.tolist(), val_df2.file_name, val_df2.width, val_df2.height):
    images_dict2 = dict(id = item[0], file_name = item[1] + '.png', width = item[2], height = item[3])
    images_list2.append(images_dict2)
```

```python
annotations_list = []
for item in zip(train_df.index.tolist(), train_df.image_number, train_df.class_id, train_df.x_min, train_df.y_min,
               train_df.box_width, train_df.box_height, train_df.iscrowd, train_df.area):
    annotations_dict = dict(id = item[0], image_id = item[1], category_id = item[2], bbox = [item[3], item[4], item[5], item[6]],
                           iscrowd = item[7], area = item[8])
    annotations_list.append(annotations_dict)
    
annotations_list2 = []
for item in zip(val_df.index.tolist(), val_df.image_number, val_df.class_id, val_df.x_min, val_df.y_min,
               val_df.box_width, val_df.box_height, val_df.iscrowd, val_df.area):
    annotations_dict2 = dict(id = item[0], image_id = item[1], category_id = item[2], bbox = [item[3], item[4], item[5], item[6]],
                           iscrowd = item[7], area = item[8])
    annotations_list2.append(annotations_dict2)
```

```python
categories_list = []
for key, value in categories_dict.items():
    temp = dict(id = key, name = value, supercategory = None)
    categories_list.append(temp)
```

```python
def make_coco(images_list, annotations_list, categories_list):
    x_ray_coco = OrderedDict()

    x_ray_coco["images"] = images_list
    x_ray_coco["annotations"] = annotations_list
    x_ray_coco["categories"] = categories_list

    if json.dumps(x_ray_coco, ensure_ascii = False, indent = "\t"):
        return x_ray_coco
```

```python
train_dict = make_coco(images_list, annotations_list, categories_list)
val_dict = make_coco(images_list2, annotations_list2, categories_list)
```

```python
with open('../data/coco/annotations/train.json', 'w', encoding="utf-8") as f1:
    json.dump(train_dict, f1, ensure_ascii=False, indent="\t")
```
